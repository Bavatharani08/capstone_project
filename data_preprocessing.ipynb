```python
n_data['Age']=n_data['Age'].fillna(n_data['Age'].median())
```

```python
n_data.isnull().sum()
```

```python
for i in n_data:
    q1 = np.quantile(d1[i],0.25)
    q3 = np.quantile(d1[i],0.75)
    iqr = q3-q1
    ll = q1 - (1.5*iqr)
    ul = q3 + (1.5*iqr)
    for row in d1[i].index:
        if d1.loc[row,i]<ll:
            d1.loc[row,i]=ll
        if d1.loc[row,i]>ul:
            d1.loc[row,i]=ul
```
